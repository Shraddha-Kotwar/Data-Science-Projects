# Splitting of Data

```python
df.head(2)
```

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
x
```

```python
y
```

# Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
enc = LabelEncoder()
```

```python
x.Gender = enc.fit_transform(x.Gender)
```

```python
x.info()
```

# Splitting of Dataset

```python
skf = StratifiedKFold(n_splits=5)
```

```python
for train_index,test_index in skf.split(x,y):
  x_train,x_test = x.iloc[train_index],x.iloc[test_index]
  y_train,y_test = y.iloc[train_index],y.iloc[test_index]
```

# Feature Scaling

```python
scale = StandardScaler()
```

```python
x_train = scale.fit_transform(x_train)
x_test = scale.fit_transform(x_test)
```

```python
x_test
```

# Model Selection

```python
log = LogisticRegression()
knn = KNeighborsClassifier()
```

# Training the model

```python
log.fit(x_train, y_train)
knn.fit(x_train, y_train)
```

# Testing the model

```python
y_knn_predict = knn.predict(x_test)
```

```python
y_log_predict = log.predict(x_test)
```

```python
new_df = pd.DataFrame({"Actual":y_test,"Predicted":y_knn_predict})
```

```python
new_df.head()
```

# Performance

```python
confusion_matrix(y_test,y_knn_predict)
```

```python
confusion_matrix(y_test,y_log_predict)
```

```python
# Accuracy = (46+22)/80 =86%
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_knn_predict)
```

```python
accuracy_score(y_test,y_log_predict)
```

```python
lis = [i for i in range(2,101) if i%2!=0]
```

```python
acc=[]
dic = {}
for i in lis:
  knn = KNeighborsClassifier(n_jobs=i)
  knn.fit(x_train, y_train)
  y_knn_pred = knn.predict(x_test)
  acc.append(accuracy_score(y_test,y_knn_pred))
  dic[i] = accuracy_score(y_test,y_knn_pred)
print(max(acc))
```
