# Train Test  Split

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=59)

model = LogisticRegression()
model.fit(x_train,y_train)
result = model.score(x_test,y_test)
print(result)
# When Random State =0, accuracy=0.9532163742690059
# When Random State =100, accuracy=0.9473684210526315
# When Random State =59, accuracy=0.9415204678362573
```

Thus ,we have seen that as the random state changes the accuracy also changes.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

for i in range(50):
  x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=i)

  model = LogisticRegression()
  model.fit(x_train,y_train)
  result = model.score(x_test,y_test)
  print(result)
```

# KFold Cross Validation

```python
from sklearn.model_selection import KFold,cross_val_score
```

```python
k = KFold(10)
```

```python
model = LogisticRegression()
result = cross_val_score(model,x,y,cv=k)
print(result)
```

```python
model = LogisticRegression()
lis=[]
arr = [5,10,20,30,40,50]
for i in arr:
  result = cross_val_score(model,x,y,cv=i)
  print(f"When k={i}",np.mean(result))
  lis.append(np.mean(result))
print(max(lis))
```

```python
#Output for above code:
'''When k=5 0.943766495885732
When k=10 0.9402882205513784
When k=20 0.9402709359605913
When k=30 0.9455165692007793
When k=40 0.9492857142857144
When k=50 0.9459090909090909
max accuracy = 0.9492857142857144'''
```

# Stratified K-fold Cross Validation

```python
from sklearn.model_selection  import StratifiedKFold
```

```python
skfold = StratifiedKFold(n_splits=5)
result = cross_val_score(model,x,y,cv=skfold)
print(np.mean(result))
```

# LOOCV- Leave One Out CV

```python
from sklearn.model_selection import LeaveOneOut
lc = LeaveOneOut()
result = cross_val_score(model,x,y,cv=lc)
print(np.mean(result))
```
